```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("amazon.csv")
df
```

```python
df.info()
df.describe()
```

```python
df.duplicated().sum()
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
df['discounted_price'].unique()
df['discounted_price']=df['discounted_price'].str.replace('₹','')
df['discounted_price']=df['discounted_price'].str.replace(',','')
df['discounted_price']=df['discounted_price'].str.replace('.','')
df['discounted_price'].unique()
df['discounted_price']=df['discounted_price'].astype(int)
```

```python
df['actual_price'].unique()
df['actual_price']=df['actual_price'].str.replace('₹','')
df['actual_price']=df['actual_price'].str.replace(',','')
df['actual_price']=df['actual_price'].str.replace('.','')
df['actual_price']=df['actual_price'].astype(int)
```

```python
df['discount_percentage'].unique()
def change_percentage(discount_percentage):
  if '%'in discount_percentage:
    return float(discount_percentage.replace('%',''))/100
  else:
      return np.nan
```

```python
df['discount_percentage']=df['discount_percentage'].apply(change_percentage)
```

```python
df.info()
```

```python
df['rating_count'].unique()
df['rating_count']=df['rating_count'].str.replace(',','')
df['rating_count']=df['rating_count'].astype(float)
```

```python
df['rating_count'].unique()
```

```python
df['rating_count'].fillna(df['rating_count'].mean(),inplace=True)
```

```python
df['rating']=df['rating'].str.replace('|','0')
df['rating']=df['rating'].astype(float)
```

```python
product_rating=df.groupby('category')['rating'].mean()
product_rating
```

```python
top_rated=df.groupby('category')['rating_count'].sum().sort_values(ascending=False).head(10)
top_rated
```

```python
plt.figure(figsize=(10,5))
plt.title('Distribution between discounted_price vs actual_price')
plt.scatter(x=df['discounted_price'],y=df['actual_price'])
plt.show()
```

```python
average_discount=df.groupby('category')['discount_percentage'].mean()
average_discount
```

```python
popular_names=df.groupby('product_name')['product_name'].value_counts().sort_values(ascending=False)
popular_names
```

```python
product_reviews=df.groupby('product_name')['review_id'].value_counts().sort_values(ascending=False)
product_reviews.head(10)
```

```python
df['discounted_price'].corr(df['rating'])
```

```python
Top_five=df.groupby('category')['rating'].count().sort_values(ascending=False).head(5)
Top_five
```

```python
df2
df2.rename(columns={'Duration (ms)':'Duration(ms)','Track Name':'Track_name','Track ID':'Track_id'},inplace=True)
```

```python
plt.figure(figsize=(10,5))
plt.title('Relationship between Popularity and Duration(ms)')
sns.scatterplot(x='Popularity', y='Duration(ms)', data=df2,c='r')
plt.show()
```

```python
highest_tracks=df2.groupby('Artist')['Track_name'].count().sort_values(ascending=False).head(10)
highest_tracks
plt.figure(figsize=(15,10))
plt.title('Highest Number of Tracks')
sns.countplot(x=df2['Artist'],order=highest_tracks.index)
plt.show()
```

```python
least_popular=df2.groupby(['Artist','Track_name'])['Popularity'].sum().sort_values(ascending=True).head(5)
least_popular
```

```python
most_popular=df2.groupby(['Artist','Track_name'])['Popularity'].sum().sort_values(ascending=False).head(5)
most_popular
```

```python
avg_popular=df2.groupby(['Artist','Track_name'])['Popularity'].mean().sort_values(ascending=False).head(5)
avg_popular
```

```python
sns.pairplot(data=df2)
plt.show()
```

```python
plt.figure(figsize=(30,50))
plt.title('Variation of Duration(ms) across various Artist')
sns.boxplot(data=df2,x='Duration(ms)',y='Artist',hue='Artist')
plt.xlabel=('Duration(ms)')
plt.ylabel=('Artist')
plt.show()
```

```python
plt.figure(figsize=(10,40))
plt.title('Popularity across various Artist')
sns.swarmplot(data=df2,x='Popularity',y='Artist',hue='Artist')
plt.xlabel=('Popularity')
plt.ylabel=('Artist')
plt.show()
```
